#Spectral Interpolation in 1D

This notebook is used to explore spectral interpolation in 1D. In all of the plots below the following function is used.

$$ f(x) = e^{-|x - x_0| / 2} $$

For each plot $x_0$ is used to center the function. This function is ploted on 3 domains of size 60, 120, and 180.



```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 60
X_small = np.arange(1. * N)
X_large = np.arange(2. * N)
X_big = np.arange(3. * N)
```

```python
def compare_plots(plot1, plot2):
    fig, axes = plt.subplots(ncols=2, figsize=(10, 5))
    ax0, ax1 = axes.flat
    
    ax0.imshow(np.abs(plot1), interpolation='none')
    ax1.imshow(np.abs(plot2), interpolation='none')
```

#Plots

The following three plots are investigated.

```python
y_small = np.exp(- np.abs(X_small - N / 2 ) / 2)
plt.plot(X_small, y_small, 'ko-')
```

```python
y_large = np.exp(- np.abs(X_large - N ) / 2.)
print y_large.shape
print X_large.shape
plt.plot(X_large, y_large, 'ko-')
```

```python
y_big = np.exp(- np.abs(X_big - (3 / 2.) * N) / 2. )
plt.plot(X_big, y_big, 'ko-')
```

#Fourier Transforms

The following three plots are the corresponding Fourier transforms of the 3 plots above. 

```python
fy_small = np.fft.fftn(y_small)
plt.plot(X_small, np.abs(fy_small), 'ko-')
```

```python
fy_large = np.fft.fftn(y_large)
plt.plot(X_large, np.abs(fy_large), 'ko-')
```

```python
fy_big = np.fft.fftn(y_big)
plt.plot(X_big, abs(fy_big), 'ko-')
```

```python
fy_pad_large = np.zeros(2 * N, dtype='complex')
fy_pad_large[:(N / 2)] = fy_small[:(N / 2)]
fy_pad_large[(N + N / 2):] =  fy_small[(N / 2):]


#fy_pad_large[:N][::2] = fy_small[:(N / 2)]
#fy_pad_large[N:][::2] =  fy_small[(N / 2):]
plt.plot(X_large, abs(fy_pad_large), 'bo-', label='Zero Padded Signal')
plt.plot(X_large, abs(fy_large), 'ko-', label='Actual Signal')
plt.legend()
```

```python
fy_pad_big = np.zeros(3 * N, dtype='complex')
fy_pad_big[:(N / 2)] = fy_small[:(N / 2)]
fy_pad_big[(2 * N + N / 2):] = fy_small[(N / 2):]
plt.plot(X_big, np.abs(fy_pad_big), 'ko-')
plt.plot(X_big, abs(fy_pad_big), 'bo-', label='Zero Padded Signal')
plt.plot(X_big, abs(fy_big), 'ko-', label='Actual Signal')
plt.legend()
```

```python
y_pad_large = np.fft.ifftn(fy_pad_large)
plt.plot(X_large, abs(y_pad_large), 'bo-', label='Zero Padded Signal')
plt.plot(X_large, abs(y_large), 'ko-', label='Actual Signal')
plt.legend()
```

```python
y_pad_big = np.fft.ifftn(fy_pad_big)
plt.plot(X_big, y_pad_big, 'ko-')
plt.plot(X_big, abs(y_pad_big), 'bo-', label='Zero Padded Signal')
plt.plot(X_big, abs(y_big), 'ko-', label='Actual Signal')
plt.legend()
```

```python
fy_pad_large = np.zeros(2 * N, dtype='complex')
fy_pad_large[:N][::2] = fy_small[:(N / 2)]
fy_pad_large[N:][::2] =  fy_small[(N / 2):]
plt.plot(X_large, abs(fy_pad_large), 'bo-', label='Zero Padded Signal')
plt.plot(X_large, abs(fy_large), 'ko-', label='Actual Signal')
plt.legend()
```

```python
plt.plot(X_large, fy_pad_large.imag, 'bo-', label='Zero Padded Signal')
plt.plot(X_large, fy_large.imag, 'ko-', label='Actual Signal')
plt.legend()
```
