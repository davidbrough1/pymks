#Coefficients and Rescaling for Simple Microstructures

This notebook is trying to understand the issue pymks is having with actually predicting the strain for a binary materia.

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import pymks
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

**Make 21 x 21 and 41 x 41 binary microstructures**

```python
L = 21
X_L = np.zeros((1, L, L), dtype='int')
X_L[:, L / 2:, :] = 1
```

```python
N = 41
X_N = np.zeros((1, N, N), dtype='int')
X_N[:, N / 2:, :] = 1
```

**Make 21 x 21 delta microstructures and strain fields** 
**Calibration influence coefficients** 

```python
from pymks.tools import draw_coeff
from pymks.datasets import make_elastic_FE_strain_delta
from pymks import MKSRegressionModel
from pymks.bases import DiscreteIndicatorBasis

elastic_modulus = (100, 80)
poissons_ratio = (0.3, 0.3)
macro_strain = 0.02

X, y = make_elastic_FE_strain_delta(elastic_modulus=elastic_modulus, poissons_ratio=poissons_ratio,
                                    size=(L, L), macro_strain=macro_strain)

basis = DiscreteIndicatorBasis(n_states=2, domain=[0, 1])
model = MKSRegressionModel(basis=basis)
model.fit(X, y)

draw_coeff(model.coeff)
```

**Make strain data from FE model for binary microstructure**

```python
from pymks.datasets.elastic_FE_simulation import ElasticFESimulation

sim = ElasticFESimulation(elastic_modulus=elastic_modulus, poissons_ratio=poissons_ratio, macro_strain=macro_strain)

sim.run(X_L)
y_L = sim.response
```

```python
from pymks.tools import draw_strains_compare

y_L_pred = model.predict(X_L)

draw_strains_compare(y_L[0], y_L_pred[0])
print 'Domain size', y_L[0].shape
print 'k = 0 frequceies', model._filter.Fkernel[0, 0, 0, :]
```

**Resize Coefficients to 41 x 41**

```python
model.resize_coeff(X_N[0].shape)

draw_coeff(model.coeff)
#print model._filter.Fkernel[0, 0, 0, :]
```

**Run FE simulation for 41 x 41 binary microstructure**

```python
sim.run(X_N)
y_N = sim.response
```

**Predict strain field with resized influence coefficients**

```python
y_N_predict = model.predict(X_N)
```

```python
from pymks.tools import draw_strains_compare

draw_strains_compare(y_N[0], y_N_predict[0])

print 'Domain size', y_N[0].shape
print 'k = 0 frequceies', model._filter.Fkernel[0, 0, 0, :]
```

```python
coeffs = model.coeff
print coeffs[0, 0, :]
```

**Generate 41 x 41 delta microstructures and strain fields**
**Fit 41 x 41 coefficients** 

```python
X_N_delta, y_N_delta = make_elastic_FE_strain_delta(elastic_modulus=elastic_modulus, poissons_ratio=poissons_ratio,
                                    size=(N, N), macro_strain=macro_strain)

model.fit(X_N_delta, y_N_delta)
```

```python
y_N_delta_predict = model.predict(X_N)

draw_strains_compare(y_N[0], y_N_delta_predict[0])

print 'Domain size', y_N[0].shape
print 'k = 0 frequceies', model._filter.Fkernel[0, 0, 0, :]
```

**Find differences in k = 0 frequencies for each of the local states**

```python
Fcoeff_diff = model._filter._real_2_frequency(model.coeff - coeffs)
print np.unique(Fcoeff_diff)
```
