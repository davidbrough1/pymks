```python
def plot_coefficients(Fcoeffs):
    fig, axes = plt.subplots(ncols=3, nrows=2, figsize=(15, 10))
    ax0, ax1, ax2, ax3, ax4, ax5 = axes.flat
    
    ax0.imshow(np.abs(Fcoeffs[..., 0]), interpolation='none')
    ax1.imshow(Fcoeffs[..., 0].imag, interpolation='none')
    ax2.imshow(Fcoeffs[..., 0].real, interpolation='none')
    ax3.imshow(np.abs(Fcoeffs[..., 1]), interpolation='none')
    ax4.imshow(Fcoeffs[..., 1].imag, interpolation='none')
    ax5.imshow(Fcoeffs[..., 1].real, interpolation='none')
    
    print np.min(Fcoeffs[..., 0]), np.max(Fcoeffs[..., 0])
    print np.min(Fcoeffs[..., 1]), np.max(Fcoeffs[..., 1])
    
    plt.show()
```

```python
def plot_response_coeffs(micro, pred0, coeff0, pred1, coeff1):
    fig, axes = plt.subplots(ncols=4, nrows=2, figsize=(20, 10))
    ax0, ax1, ax2, ax3, ax4, ax5, ax6, ax7 = axes.flat
    
    ax0.imshow(micro[0], interpolation='none')
    ax1.imshow(pred0[0], interpolation='none')
    ax2.imshow(coeff0[..., 0], interpolation='none')
    ax3.imshow(coeff0[..., 1], interpolation='none')
    ax4.imshow(micro[0], interpolation='none')
    ax5.imshow(pred1[0], interpolation='none')
    ax6.imshow(coeff1[..., 0], interpolation='none')
    ax7.imshow(coeff1[..., 1], interpolation='none')
    plt.show()
```

```python
from pymks.tools import draw_coeff
from pymks.datasets import make_elastic_FE_strain_delta
from pymks import MKSRegressionModel
from pymks.bases import DiscreteIndicatorBasis
```

```python
elastic_modulus = (100, 80)
poissons_ratio = (0.3, 0.3)
macro_strain = 0.02

L = 21

X_L, y_L = make_elastic_FE_strain_delta(elastic_modulus=elastic_modulus, poissons_ratio=poissons_ratio,
                                    size=(L, L), macro_strain=macro_strain)
```

```python
X_2L, y_2L = make_elastic_FE_strain_delta(elastic_modulus=elastic_modulus, poissons_ratio=poissons_ratio,
                                    size=(2 * L, 2 * L), macro_strain=macro_strain)
```

```python
X_3L, y_3L = make_elastic_FE_strain_delta(elastic_modulus=elastic_modulus, poissons_ratio=poissons_ratio,
                                    size=(3 * L, 3 * L), macro_strain=macro_strain)
```

#Influence Coefficients

All of the images in this notebook come from a 2 phase microstructure with a contrast of $\frac{120}{80}$. 

Each row of the plots below represent a different representation of an influence ceofficients in Fourier space (what is referred to as $\gamma_k$ in this [paper](http://dx.doi.org/10.1016/j.actamat.2010.01.007)).

The left column is the abosolute value of the coefficient values, the middle coloumn is the imaginary components of the coefficients, and the right column is the real components of the coefficients.

```python
basis = DiscreteIndicatorBasis(n_states=2, domain=[0, 1])
model = MKSRegressionModel(basis=basis)
model.fit(X_L, y_L)
plot_coefficients(model._filter.Fkernel[0])
```

```python
np.random.seed(99)
X_L_test = np.random.randint(2, size=(1, L, L))
X_2L_test = np.random.randint(2, size=(1, 2 * L, 2 * L))
X_3L_test = np.random.randint(2, size=(1, 3 * L, 3 * L))
```

```python
model.fit(X_L, y_L)
model.resize_coeff(2)
coeff0 = model.coeff
y_2L_pred0 = model.predict(X_2L_test)
model.fit(X_2L, y_2L)
coeff1 = model.coeff
y_2L_pred1 = model.predict(X_2L_test)

plot_response_coeffs(X_2L_test, y_2L_pred0, coeff0, y_2L_pred1, coeff1)
```

```python
model.fit(X_L, y_L)
model.resize_coeff(3)
coeff2 = model.coeff
y_3L_pred2 = model.predict(X_3L_test)
model.fit(X_3L, y_3L)
coeff3 = model.coeff
y_3L_pred3 = model.predict(X_3L_test)

plot_response_coeffs(X_3L_test, y_3L_pred2, coeff2, y_3L_pred3, coeff3)
```
